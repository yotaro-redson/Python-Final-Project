```python
import nltk as nltk
from nltk.tokenize import word_tokenize
import re
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
nltk.download('punkt')


```

```python

stemmer = SnowballStemmer("english")
def tokenize_and_stem(text):
    if isinstance(text, str):  # Check if the input is a string
        # Tokenize by sentence, then by word
        tokens = [word for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
        
        # Filter out raw tokens to remove noise
        filtered_tokens = [token for token in tokens if re.search('[a-zA-Z]', token)]
        
        # Stem the filtered tokens
        stems = [stemmer.stem(t) for t in filtered_tokens]
        
        return stems
    else:
        return []  # Return an empty list if the input is not a string

plot['cleaned_overview']= plot['overview'].apply(tokenize_and_stem)
```

```python
def clean_and_stem(text):
    stemmer = SnowballStemmer("english")
    if isinstance(text,str):
        text = re.sub('[^a-zA-Z0-9\s]', '', text)
        tokens = word_tokenize(text)

        stop_words =set(stopwords.words('english'))
        tokens = (word for word in tokens if word.lower() not in stop_words)

        stemmed_tokens = (stemmer.stem(tokens) for token in tokens)
        return stemmed_tokens
    else:
        return []
# Apply the function to the DataFrame
plot['cleaned_overview'] = plot['overview'].apply(clean_and_stem)

```
